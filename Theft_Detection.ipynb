```python
import pandas as pd
import numpy as np

# Create a date range for 84 days (12 weeks)
dates = pd.date_range(start="2024-09-01", periods=84)

# Total number of consumers
num_consumers = 380

# Generate unique consumer IDs and smart meter numbers
consumer_ids = [f'CID{str(i).zfill(3)}' for i in range(1, num_consumers + 1)]
smart_meter_ids = [f'SM{str(i).zfill(3)}' for i in range(1, num_consumers + 1)]

# Initialize dictionary to store datasets by week
all_datasets = {f'Week {i+1}': [] for i in range(12)}

# Initialize a list to store total consumption per week
weekly_total_consumption = []

# Function to calculate non-zero counts
def non_zero_count(df, column_name):
    return (df[column_name] != 0).sum()

# Function to calculate total consumption for each day
def calculate_total_consumption(df):
    total_consumption = {
        'Week': df['Week'].iloc[0],
        'Date': 'Total Consumption',
        'Consumer Identification Number': '',
        'Smart Meter Number': '',
        '1000 kWh': df['1000 kWh'].sum() / 1000,  # Convert to MWh
        '2000 kWh': df['2000 kWh'].sum() / 1000,  # Convert to MWh
        '3000 kWh': df['3000 kWh'].sum() / 1000,  # Convert to MWh
        '4000 kWh': df['4000 kWh'].sum() / 1000,  # Convert to MWh
        '5000 kWh': df['5000 kWh'].sum() / 1000   # Convert to MWh
    }
    return total_consumption

# Populate data for each date
for idx, date in enumerate(dates):
    week = (idx // 7) + 1  # Determine the week number
    data = []
    for cid, sm in zip(consumer_ids, smart_meter_ids):
        consumption = {
            'Week': f'Week {week}',
            'Date': date,
            'Consumer Identification Number': cid,
            'Smart Meter Number': sm,
            '1000 kWh': np.random.randint(100, 1000) if np.random.rand() > 0.3 else 0,
            '2000 kWh': np.random.randint(100, 2000) if np.random.rand() > 0.5 else 0,
            '3000 kWh': np.random.randint(100, 3000) if np.random.rand() > 0.7 else 0,
            '4000 kWh': np.random.randint(100, 4000) if np.random.rand() > 0.9 else 0,
            '5000 kWh': np.random.randint(100, 5000) if np.random.rand() > 0.95 else 0
        }
        data.append(consumption)

    # Convert to DataFrame
    df = pd.DataFrame(data)

    # Calculate total consumers for non-zero entries
    total_consumers = {
        'Week': f'Week {week}',
        'Date': 'Total Consumers',
        'Consumer Identification Number': '',
        'Smart Meter Number': '',
        '1000 kWh': non_zero_count(df, '1000 kWh'),
        '2000 kWh': non_zero_count(df, '2000 kWh'),
        '3000 kWh': non_zero_count(df, '3000 kWh'),
        '4000 kWh': non_zero_count(df, '4000 kWh'),
        '5000 kWh': non_zero_count(df, '5000 kWh')
    }
    total_consumers_df = pd.DataFrame([total_consumers])

    # Calculate total consumption for the current day
    total_consumption_df = pd.DataFrame([calculate_total_consumption(df)])

    # Concatenate the total_consumers_df and total_consumption_df to the main DataFrame
    df = pd.concat([df, total_consumers_df, total_consumption_df], ignore_index=True)

    # Append the DataFrame to the corresponding week list in the dictionary
    all_datasets[f'Week {week}'].append(df)

    # Add the daily total consumption to the weekly summary
    if idx % 7 == 6:  # End of the week
        weekly_total = calculate_total_consumption(pd.concat(all_datasets[f'Week {week}']))
        weekly_total_consumption.append(weekly_total)

# Print the datasets for each week
for week, datasets in all_datasets.items():
    print(f"Datasets for {week}:")
    for dataset in datasets:
        print(dataset)
        print("\n" + "-"*100 + "\n")

# Create and print the weekly total consumption dataset
weekly_total_df = pd.DataFrame(weekly_total_consumption)
print("Weekly Total Consumption Dataset:")
print(weekly_total_df)
```
